```python
import numpy as np
import pywt
import pywt.data
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread('/content/drive/MyDrive/FWT_data/lena.bmp')
```

```python
coeffs2 = pywt.dwt2(img, 'db3', mode = 'periodization') # level 1 DWT
cA, (cH, cV, cD) = coeffs2 # exctracting coeffs
```

```python
imgr = pywt.idwt2(coeffs2, 'db3', mode = 'periodization')
imgr = np.uint8(imgr)
```

```python
plt.figure(figsize = (30, 30))
```

```python
plt.subplot(2, 2, 1)
plt.imshow(cA, cmap = plt.cm.gray)
plt.title('Approximation coefficients (cA)')

plt.subplot(2, 2, 2)
plt.imshow(cH, cmap = plt.cm.gray)
plt.title('Horizontal detailed coefficients (cH)')

plt.subplot(2, 2, 3)
plt.imshow(cV, cmap = plt.cm.gray)
plt.title('Vertical detailed coefficients (cV)')

plt.subplot(2, 2, 4)
plt.imshow(cD, cmap = plt.cm.gray)
plt.title('Diagonal detailed coefficients (cD)')

plt.subplots_adjust(hspace = 3, wspace = 3)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.imshow(imgr, cmap = plt.cm.gray)
plt.title('Reconstructed image')
```
