### **Multi-level Discrete Wavelet Transform & Inverse Transform**

```python
import pywt
import matplotlib.pyplot as plt
```

```python
# x = [3, 7, 1, 1, -2, 5,4, 6, 6, 4, 5, -2, 1, 1, 7, 3] #signal
x = [-2, 1, 4, 2, -1]
coeffs = pywt.wavedec(x, 'db1', level = 2, mode = 'periodic') # DWT
cA2, cD2, cD1 = coeffs
y = pywt.waverec(coeffs, 'db1', mode = 'periodic') # IDWT
```

```python
# Plotting
plt.figure(figsize=(5, 8))

# Plotting the original signal x as vertical lines with maroon blobs
plt.subplot(5, 1, 1, frameon=False)
plt.vlines(range(len(x)), 0, x, color='gray', linewidth=2)
plt.scatter(range(len(x)), x, color='maroon', s=30)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.title('Original Signal (x)')
plt.xticks([])  # Remove x-axis ticks

# Plotting the cA2 coefficients
plt.subplot(5, 1, 2, frameon=False)
plt.vlines(range(len(cA2)), 0, cA2, color='gray', linewidth=2)
plt.scatter(range(len(cA2)), cA2, color='maroon', s=30)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.title('Approximation Coefficients (cA2)')
plt.xticks([])  # Remove x-axis ticks

# Plotting the cD2 coefficients
plt.subplot(5, 1, 3, frameon=False)
plt.vlines(range(len(cD2)), 0, cD2, color='gray', linewidth=2)
plt.scatter(range(len(cD2)), cD2, color='maroon', s=30)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.title('Detail Coefficients (cD2)')
plt.xticks([])  # Remove x-axis ticks

# Plotting the cD1 coefficients
plt.subplot(5, 1, 4, frameon=False)
plt.vlines(range(len(cD1)), 0, cD1, color='gray', linewidth=2)
plt.scatter(range(len(cD1)), cD1, color='maroon', s=30)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.title('Detail Coefficients (cD1)')
plt.xticks([])  # Remove x-axis ticks

# Plotting the reconstructed signal y as vertical lines with maroon blobs
plt.subplot(5, 1, 5, frameon=False)
plt.vlines(range(len(y)), 0, y, color='gray', linewidth=2)
plt.scatter(range(len(y)), y, color='maroon', s=30)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Add horizontal line at y=0
plt.title('Reconstructed Signal (y)')
plt.xticks([])  # Remove x-axis ticks

plt.tight_layout()
plt.show()
```

```python
print('cA2 = ', cA2)
print('cD2 = ', cD2)
print('cD1 = ', cD1)
print('Reconstructed signal = ', y)
```
