## **Multi-level wavelet decomposition & reconstruction of 2D signals**

```python
import numpy as np
import pywt
import pywt.data
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# img = mpimg.imread('/content/drive/MyDrive/FWT_data/w256a.bmp')
# img.shape
```

```python
def convert_to_grayscale(input_image_path):
    # Open the image using PIL
    img = Image.open(input_image_path)

    # Convert the image to grayscale
    gray_img = img.convert('L')

    # Save the grayscale image
    # gray_img.save(output_image_path)

    return gray_img
```

```python
img = convert_to_grayscale('/content/drive/MyDrive/FWT_data/w256a.bmp')
img.size
```

```python
C = pywt.wavedec2(img, 'db5', mode = 'periodization', level = 1) # level 1 DWT
imgr = pywt.waverec2(C, 'db5', mode = 'periodization') # level 1 IDWT
imgr = np.uint8(imgr)

cA1 = C[0]
(cH1, cV1, cD1) = C[-1]

plt.figure(figsize = (20, 20))

plt.subplot(2, 2, 1)
plt.imshow(cA1, cmap = plt.cm.gray)
plt.title('Approximation coefficients (cA1)')

plt.subplot(2, 2, 2)
plt.imshow(cH1, cmap = plt.cm.gray)
plt.title('Horizontal detailed coefficients (cH1)')

plt.subplot(2, 2, 3)
plt.imshow(cV1, cmap = plt.cm.gray)
plt.title('Vertical detailed coefficients (cV1)')

plt.subplot(2, 2, 4)
plt.imshow(cD1, cmap = plt.cm.gray)
plt.title('Diagonal detailed coefficients (cD1)')
```

```python
C = pywt.wavedec2(img, 'db5', mode = 'periodization', level = 2) # level 2 DWT
imgr = pywt.waverec2(C, 'db5', mode = 'periodization') # level 2 IDWT
imgr = np.uint8(imgr)

cA2 = C[0]
(cH1, cV1, cD1) = C[-1]
(cH2, cV2, cD2) = C[-2]

plt.figure(figsize = (20, 20))

plt.subplot(2, 2, 1)
plt.imshow(cA2, cmap = plt.cm.gray)
plt.title('Approximation coefficients (cA2)')

plt.subplot(2, 2, 2)
plt.imshow(cH2, cmap = plt.cm.gray)
plt.title('Horizontal detailed coefficients (cH2)')

plt.subplot(2, 2, 3)
plt.imshow(cV2, cmap = plt.cm.gray)
plt.title('Vertical detailed coefficients (cV2)')

plt.subplot(2, 2, 4)
plt.imshow(cD2, cmap = plt.cm.gray)
plt.title('Diagonal detailed coefficients (cD2)')
```

```python
C = pywt.wavedec2(img, 'db5', mode = 'periodization', level = 3) # level 3 DWT
imgr = pywt.waverec2(C, 'db5', mode = 'periodization') # level 3 IDWT
imgr = np.uint8(imgr)

cA3 = C[0]
(cH1, cV1, cD1) = C[-1]
(cH2, cV2, cD2) = C[-2]
(cH3, cV3, cD3) = C[-3]

plt.figure(figsize = (20, 20))

plt.subplot(2, 2, 1)
plt.imshow(cA3, cmap = plt.cm.gray)
plt.title('Approximation coefficients (cA3)')

plt.subplot(2, 2, 2)
plt.imshow(cH3, cmap = plt.cm.gray)
plt.title('Horizontal detailed coefficients (cH3)')

plt.subplot(2, 2, 3)
plt.imshow(cV3, cmap = plt.cm.gray)
plt.title('Vertical detailed coefficients (cV3)')

plt.subplot(2, 2, 4)
plt.imshow(cD3, cmap = plt.cm.gray)
plt.title('Diagonal detailed coefficients (cD3)')
```

```python
arr, coeff_slices = pywt.coeffs_to_array(C)
plt.figure(figsize = (20, 20))
plt.imshow(arr, cmap = plt.cm.gray)
plt.title('All wavelet coefficients upto level 3')
```

```python
plt.figure()
plt.imshow(imgr, cmap = plt.cm.gray)
plt.title('Reconstructed image')
```

```python
compressed = arr[coeff_slices[0]]
plt.figure()
plt.imshow(compressed, cmap = plt.cm.gray)
plt.title('Compressed image')
# plt.savefig('compressed.png')
```
