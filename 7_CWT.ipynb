```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 200)
signal = np.cos(2*np.pi*7*t) + np.real(np.exp(-7*(t-0.4)**2) * np.exp(1j*2*np.pi*2*(t-0.4)))
scales = np.arange(1, 31)
coeff, freqs = pywt.cwt(signal, scales, 'gaus1')
```

```python
plt.figure(figsize = (15, 10))
plt.plot(t, signal)
plt.grid(color = 'gray', linestyle = ':', linewidth = 0.5)
```

```python
# Plotting scaleogram

plt.figure(figsize = (15, 10))

plt.imshow(abs(coeff), extent =[0, 200, 30, 1], interpolation = 'bilinear', cmap = 'gnuplot2', aspect = 'auto', vmax = abs(coeff).max(), vmin = -abs(coeff).max())

plt.gca().invert_yaxis()
plt.yticks(np.arange(1, 31, 1))
plt.xticks(np.arange(0, 201, 10))
plt.show()
```
